### Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Setting global variables

```python
# Setting global variable for the ipynb file
data_dir = '../data'
```

### Importing the dataset

```python
# Loading the training and testing dataset
train_df = pd.read_csv(os.path.join(data_dir, 'train.csv'))
test_df = pd.read_csv(os.path.join(data_dir, 'test.csv'))
```

### Data Description

Data quality assessment is a crucial step in any data analysis or machine learning project. This code snippet utilizes the `info()` method to provide essential information about the structure of the datasets `train_df` and `test_df`. This includes details such as the number of entries, data types, and presence of any missing values. By examining this information, we gain insights into the completeness and consistency of the datasets, enabling us to make informed decisions for subsequent data preprocessing and analysis steps.


```python
# Check data quality
print(train_df.info())
print(test_df.info())
```

This code snippet provides an overview of the training dataset (`train_df`) and the testing dataset (`test_df`). 
- `.shape` returns the dimensions of the dataset, indicating the number of rows and columns.
- `.head()` displays the first few rows of the dataset, offering a glimpse of the data's structure and content.
- `.describe()` generates descriptive statistics, including measures of central tendency and dispersion, for numerical columns in the dataset. 

By examining these aspects, we gain insights into the dataset's size, structure, and statistical distribution, facilitating further exploration and analysis.

```python
# Describe the shape, head, and summary of the training dataset
print(train_df.shape)
print(train_df.head())
print(train_df.describe())
```

```python
# Describe the shape, head, and summary of the test dataset
print(test_df.shape)
print(test_df.head())
print(test_df.describe())
```

```python
# Check for missing values
print("\nData quality check:")
print(train_df.isnull().sum())
print(test_df.isnull().sum())
```

```python
# Data distribution visualization
plt.figure(figsize=(10, 6))
sns.histplot(train_df['age'], bins=20, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot to visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=train_df['age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()
```
